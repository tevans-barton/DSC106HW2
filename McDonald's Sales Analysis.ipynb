```python
import pandas as pd
import numpy as np
import matplotlib as plt
from copy import deepcopy
```

```python
daily = pd.read_csv('daily_sales.csv')
monthly = pd.read_csv('monthly_sales.csv')
```

```python
average_monthly = deepcopy(monthly)
```

```python
average_monthly['Month'] = average_monthly['Month, Year'].str.slice(0,3)
```

```python
average_monthly = average_monthly.groupby('Month').mean()
average_monthly = average_monthly.reindex(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov','Dec'])
average_monthly
```

```python
average_monthly.to_csv('average_monthly.csv')
```

```python
hamburger_monthly = average_monthly.drop(['CF-NE', 'CF-SW', 'CF-NW', 'CF-SE', 'CF-C', 'FF-NE', 'FF-NW', 'FF-SE', 'FF-SW', 'FF-C'], axis = 1)
```

```python
hamburger_monthly.to_csv('average_hamburger_monthly.csv')
```

```python
hamburgers = monthly.drop(['CF-NE', 'CF-SW', 'CF-NW', 'CF-SE', 'CF-C', 'FF-NE', 'FF-NW', 'FF-SE', 'FF-SW', 'FF-C'], axis = 1)
```

```python
hamburgers['Month'] = hamburgers['Month, Year'].str.slice(0,3)
```

```python
hamburgers['Total Sold'] = hamburgers['HM-NE'] + hamburgers['HM-SW'] + hamburgers['HM-NW'] + hamburgers['HM-SE'] + hamburgers['HM-C']
```

```python
hamburgers_total_temp = hamburgers.drop(['HM-NE', 'HM-SW', 'HM-NW', 'HM-C', 'HM-SE'], axis = 1)
```

```python
hamburgers_total_temp['Year'] = hamburgers_total_temp['Month, Year'].str.slice(4,6)
```

```python
hamburgers_total = pd.pivot_table(hamburgers_total_temp, values = 'Total Sold', index = 'Month', columns = 'Year')
hamburgers_total = hamburgers_total.reindex(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov','Dec'])
```

```python
hamburgers_total.to_csv('hamburgers_total.csv')
```
